### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
def findDogs_and_Chickens(heads,legs):
    flag = True
    if legs%2 != 0:
        flag = False
    if legs < 2 * heads or legs > 4 * heads:
        flag = False         

    if flag:
        dogs = (legs - 2*(heads)) / 2
        chickens = heads - dogs 
        return dogs, chickens
    else:
        return 'Please enter details again'
findDogs_and_Chickens(5,18)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
def SumOfSquares(n):
    sum = 0
    for num in range(1,n+1):
        sum += num**2
    return sum
SumOfSquares(5)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
def ArithmeticSeries(t0,t1,n):
    d = t1 - t0
    tN = t0 + (n-1) * d
    return tN

ArithmeticSeries(0,2,5)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here
def FractionalSum(a,b,c,d):
    numerator = a*d + b*c
    denominator = b*d
    return f'{numerator}/{denominator}'
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Write your code here
def FindGlassesOfMilk(tankH,tankL,tankB,glassH,glassR):
    container_volume = tankB * tankH * tankL
    glass_volume =  3.14 * (glassR**2) * glassH

    no_of_glasses = container_volume // glass_volume
    return no_of_glasses

FindGlassesOfMilk(20,20,20,10,3)
````

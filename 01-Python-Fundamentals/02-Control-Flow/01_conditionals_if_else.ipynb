### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Write code here
def inHandSalary(CTC):
    HRA_deduction = CTC * 0.1
    DA_deduction = CTC * 0.05
    PF_deduction = CTC * 0.03
    inHand = CTC - HRA_deduction - DA_deduction - PF_deduction

    if inHand < 5:
        return inHand/12
    elif 5 < inHand < 10:
        return (inHand - inHand*0.1)/12
    elif 10 < inHand < 20:
        return (inHand - inHand*0.2)/12
    elif inHand > 20:
        return (inHand - inHand*0.3)/12
        
inHandSalary(20)
````

### `Problem 5` - Exercise 12: Display Fibonacci series up to 10 terms.

*Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34*

````python
# Write code here
def fibb(n):
    a = 0
    b = 1
    c = a + b
    print(0,1,end=' ')
    for num in range(n-2):
        print(c,end=' ')
        a = b
        b = c
        c = a + b

fibb(10)
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
# Write code here
def factorial(n):

    if n < 0:
        return None
    elif n == 0:
        return 1
    elif n == 1:
        return 1
    fctrl = 1

    for i in range(1,n+1):
        fctrl *= i
    return fctrl

factorial(5)
````

### `Problem 7` - Reverse a given integer number.

Example:

`Input:`

```bash
76542
```

`Output:`

```bash
24567
```

````python
# Write code here
def reverseNumber(n):
    rev = 0

    while n > 0:
        rev = 10 * rev + n % 10
        n //= 10
    return rev

reverseNumber(24567)
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# Write code here
def sum_skip_5(N):
    i = 1
    sum = 0 

    while i <= N:
        if i  %  5 == 0:
            i += 1
        if i + sum > 300:
            break
        
        sum += i
        i += 1
    return sum

sum_skip_5(30)
````

###`Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# This code calculates the euclidean distance (rounded off to the nearest integer) 
# between the robot's current position and the robot's starting position

def euclideanDistance(x1,y1,x2,y2):
    return ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5

def getCurrentCoord(U,D,L,R):
    x = R - L
    y = U - D
    return x, y

def getInitialCoord():
    coord = input("Enter in format 'x, y': ").split(',')
    x, y = map(int, coord)
    return x, y

def main():
    
    print("Welcome to STALK THE ROBOT (umm yaa!)")
    print("let your robot wander off by inputting the Direction and Distance")
    print("wanna go up by three points? input -> 'U, 3'")
    print("wanna go left by one point? input -> 'L, 1'")
    print("wanna stopp? input -> '!'")
    print("you are controlling enough already to understand the moves!")
    print('let the robot roll! Enter the initial drop point \n')

    moves = {'U': 0, 'D': 0, 'L': 0, 'R': 0}
    x1, y1 = getInitialCoord() 

    print(f'Robot dropped at {x1, y1}! \n')

    while True:
        inp = input('Enter Direction, Distance eg: "R, 4" or "!" to stop: ').split(',')
        if inp[0] == '!':
            break
        else:
            direction, distance = inp[0].strip(), int(inp[1])
            moves[direction] = moves.get(direction) + distance
            x,y = getCurrentCoord(moves['U'],moves['D'],moves['L'],moves['R'])

            print(f'Robot is currently at {x,y}\n')

    x2, y2 = getCurrentCoord(moves['U'],moves['D'],moves['L'],moves['R'])
    
    distance = euclideanDistance(x1,y1,x2,y2)

    print(f'The final distance is {round(distance)} from {x1,y1}')

main()
````

###`Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
# Write code here
def getArmstrongsFrom(start: int, end: int) -> None:
    for num in range(start, end+1):
        copy = num
        temp = 0
        while copy != 0:
            temp += (copy%10)**3
            copy //= 10
        if temp == num:
            print(num, end=' ')

getArmstrongsFrom(0,500)
````

````python
# Write code here
def getAngleBetween(H: int ,M: int): 
    min_hand_position = M * 6
    hour_hand_position = H * 30 + (M * 0.5)

    raw_angle = abs(hour_hand_position - min_hand_position)
    final_angle = min(raw_angle, 360-raw_angle)
    return int(final_angle)

getAngleBetween(3,59)

````
